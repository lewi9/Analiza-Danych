### Zadnie 2: Prezentacja danych w formie tabelarycznej [10 pkt]

Przygotuj trzy tablice szeregów rodzielczych przedziałowych częstości wskaźnika GDP growth (annual %) dla poszczególnych lat.

Pierwsza tablica (`recession_table_1`) powinna zawierać szeregi skonstruowane w obarciu o całość danych.

Indeks:
 - Year (wszystkie lata)
 
Kolumny:
 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex

Druga tablica (`recession_table_2`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego z rejonów geograficznych.

Indeks:
 - 1 poziomu: Region Name
 - 2 poziomu: Year (wszystkie lata dla każdego z rejonów)
 
Kolumny:
 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex

Trzecia tablica (`recession_table_3`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego stopnia zamożności.

Indeks:
 - 1 poziomu: Income Group
 - 2 poziomu: Year (wszystkie lata dla każdej z grup)
 
Kolumny:
 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex
 
Wyświetl przygotowane tablice w taki sposób aby zawarte w tablicy liczby (za wyjątkiem lat i granic przedziałów w nagłówkach kolumn i wierszy) wyświetlały się z dokładnością do 2 miejsca po przecinku.

```python
# Całość danych
# YOUR CODE HERE
tuples = [(x,x+5) for x in range(-20,35,5)]
bins = pd.IntervalIndex.from_tuples(tuples)
recession_table_1 = pd.DataFrame()
df = pd.DataFrame()
for i in range(2006,2012):
    df[f"{i}"] = pd.cut(recession_data_1[f"{i}"], bins)
for i in range(2006,2012):
    recession_table_1 = pd.concat([recession_table_1,df.groupby(f"{i}").count().T])
recession_table_1 = recession_table_1.iloc[0::5]/len(recession_data_1)
recession_table_1.index=[2006,2007,2008,2009,2010,2011]
recession_table_1.index.name = "Year"

# Oświadczamy, że nie zwróciliśmy należytej uwagi na zajęciach oraz niezauważliśmy dołączego pliku
# recession_table_1.pkl, dlatego sami go stworzyliśmy w oparciu o tabelę recession_table_1
# i na tym oparliśmy nasze dalsze działania.

recession_table_1.to_pickle("recession_table_1.pkl")

dfStyled = recession_table_1.style.format(precision=2)
dfi.export(dfStyled, "recession_table_1.png", chrome_path=None)
display(dfStyled)


```

```python
# Podział na rejony
# YOUR CODE HERE
tuples = [(x,x+5) for x in range(-20,35,5)]
bins = pd.IntervalIndex.from_tuples(tuples)

names = [x for x in bins]

recession_table_2 = pd.DataFrame()
df = pd.DataFrame()
df.index = recession_data_1.index

for region in recession_data_1["Region Name"].unique():
    for i in range(2006,2012):
        df[(f"{region}",f"{i}")] = pd.cut(recession_data_1[recession_data_1["Region Name"] == region][f"{i}"], bins)

for column in df.columns:
    length = len(df.groupby(column).count().T)
    
    recession_table_2 = pd.concat([recession_table_2, df.groupby(column).count().T.sort_values(by=names, ascending = False)/len(recession_data_1[recession_data_1["Region Name"] == column[0]])])
    
recession_table_2 = recession_table_2.iloc[0::length]
recession_table_2.index = pd.MultiIndex.from_tuples(df.columns, names=["Region Name", "Year"])
dfStyled = recession_table_2.style.format(precision=2)
dfi.export(dfStyled, "recession_table_2.png", fontsize=8)
display(dfStyled)
```

```python
# Podział pod względem stopnia zamożności
# YOUR CODE HERE
tuples = [(x,x+5) for x in range(-20,35,5)]
bins = pd.IntervalIndex.from_tuples(tuples)

names = [x for x in bins]

recession_table_3 = pd.DataFrame()
df = pd.DataFrame()
df.index = recession_data_1.index

for region in recession_data_1["Income Group"].unique():
    for i in range(2006,2012):
        df[(f"{region}",f"{i}")] = pd.cut(recession_data_1[recession_data_1["Income Group"] == region][f"{i}"], bins)

for column in df.columns:
    length = len(df.groupby(column).count().T)
    
    recession_table_3 = pd.concat([recession_table_3, df.groupby(column).count().T.sort_values(by=names, ascending = False)/len(recession_data_1[recession_data_1["Income Group"] == column[0]])])
    
recession_table_3 = recession_table_3.iloc[0::length]
recession_table_3.index = pd.MultiIndex.from_tuples(df.columns, names=["Income Group", "Year"])
dfStyled = recession_table_3.style.format(precision=2)
dfi.export(dfStyled, "recession_table_3.png", fontsize=10)
display(dfStyled)
```

```python
# YOUR CODE HERE
bins = np.arange(-25,45,5)
def ploter(ax, y, title):
    ax.set_facecolor("white")
    sns.histplot(y["2006"], color="slategray",     ax = ax, element="poly", fill = False, stat = "probability", bins=bins, label = "2006")
    sns.histplot(y["2007"], color="magenta",       ax = ax, element="poly", fill = False, stat = "probability", bins=bins, label = "2007")
    sns.histplot(y["2008"], color="darkgoldenrod", ax = ax, element="poly", fill = False, stat = "probability", bins=bins, label = "2008")
    sns.histplot(y["2009"], color="crimson",       ax = ax, element="poly", fill = False, stat = "probability", bins=bins, label = "2009")
    sns.histplot(y["2010"], color="royalblue",     ax = ax, element="poly", fill = False, stat = "probability", bins=bins, label = "2010")
    sns.histplot(y["2011"], color="darkgreen",     ax = ax, element="poly", fill = False, stat = "probability", bins=bins, label = "2011")
    ax.grid(True)
    ax.set_xlim([-20,35])
    ax.set_ylim([0,1])
    ax.set_title(title)
    ax.set_ylabel("częstość")
    ax.set_xlabel("wskaźnik GPD growth (annual %)")
    ax.legend(title="Year")
    
fig, ax = plt.subplots(2,5,figsize=(18,10), facecolor="white")

ploter(ax[0,0], recession_data_1, "All data")
ploter(ax[0,1], recession_data_1[recession_data_1["Income Group"] == "Low"],          "Income Group - Low")
ploter(ax[0,2], recession_data_1[recession_data_1["Income Group"] == "Lower middle"], "Income Group - Lower middle")
ploter(ax[0,3], recession_data_1[recession_data_1["Income Group"] == "Upper middle"], "Income Group - Upper middle")
ploter(ax[0,4], recession_data_1[recession_data_1["Income Group"] == "High"],         "Income Group - High")

region_names = recession_data_1["Region Name"].unique()

for i in range(5):
    ploter(ax[1,i],  recession_data_1[recession_data_1["Region Name"] == region_names[i]], f"Region Name - {region_names[i]}")

fig.suptitle("Diagramy częstości")
fig.savefig("frequency_diagrams.png")
```

#### YOUR ANSWER HERE

#### Diagramy Częstości
Wyraźnie można zaobserwować, jakie wartości pojawiały się najczęściej, jakie najrzadziej dla konkrentych serii, łatwo porównać ze sobą częstości różnych serii dla konkretnych wartości. Problemem jest to, że przy dużej ilości serii stają się mniej czytelne. Dla większej liczby niż 10 serii, naprawdę trudno byłoby się rozczytać. Ponadto jeśli chcemy sprawdzać częstość czegoś w zależności od jakieś współzmiennej, to potrzeba zrobić dużą liczbę wykresów, ponieważ na jednym byłoby to zbyt nadźgane (gdybyśmy np. dobierali styl linii w zależności od regionu...). Dane, które wizualizowaliśmy poniżej na trzech wykresach, w tym przypadku potrzebowały aż 10.

#### Boxploty
Łatwo można porównać wariancje różnych serii (wielkość pudełka), łatwy i przyjemny dostęp do takich danych jak mediana, kwartyle. Obrazuje wartości odstające. Generalnie bardzo dobrze reprezentuje podstawowe statystyki opisowe. Niestety nie mamy tak prostego dostępu do wartości konkretnych częstości, tzn. możemy przypuszczać gdzie następuje kumulacja wartości na podstawie statystyk opisowych, ale jesteśmy przekonani, że istnieją zestawy danych, dla których będzie to zwodnicze. Tak więc problem się pojawi wtedy, kiedy będziemy potrzebowali znać dokładne położenie danych (dlatego można np. łączyć boxploty ze stripplotami)

#### Violinploty
Czerpią z boxplotów, właściwie tutaj też mamy pudełka, które reprezentują statystyki opisowe, tylko obudowane są otoczką, która reprezentuje gęstość poszczególnych wartości, tak więc nie ma już tego problemu co w samych boxplotach, lepiej znamy koncentracje wartości. Niestety wartości odstające drastycznie potrafią sprawić, aby wykres był mniej czytelny. Może także pojawić się problem, aby porównać konkretne częstości w różnych seriach, jednak diagramy częstotliwości ilustrują to sporo lepiej, natomiast poglądowo violinploty są zupełnie wystarczające. Pudełko też jest naszym zdnaiem mniej czytelne niż w przypadku boxplota. Violinploty stanowią naszym staniem kompromis pomiędzy boxplotem, a diagramem częstotliwości, jednak nie przypadły nam do gustu, nie są aż tak czytelne dla nas. 

#### Swarmploty
Bardzo dobrze obrazują rozkład wartości, ich punkty koncentracji oraz miejsca, gdzie następują przerzedzenia, dobrze też ilustrują dysproporcje pomiędzy grupami dla różnej wartości współzmiennej (np. mniejsza liczność krajów z Oceanii niż z Afryki). Mamy jednak obawy, że dla dużej ilości danych wykresy mogłyby na siebie zachodzić tak jak np. 2010 rok, dane z grupy ekonomicznej High zakrywają te z Upper middle. Dużo łatwiej nam też zobaczyć kształt niż w przypadku violinplot, jednak ciężej jest nam odczytać statystyki opisowe. 

#### Stripploty
Naszym zdaniem fenomenalnie widać odstawanie pewnych wartości na tle innych, z parametrem alpha można też łatwo ocenić liczność poszczególnych kategorii (tych podanych na legendzie). Można mieć lepszą intuicje do ocenienia pewnych wartości statystyk opisowych w porównaniu do swarmplotów, jednak dalej nie dorównują w tym poziomie boxplotom. Problem może pojawić się, jeśli mamy ogromną ilość danych z identyczną wartością. Wtedy, aby móc dobrze wyróżnić poziom koncentracji w tym miejscu należałoby zmniejszyć parametr alpha, ale wtedy też pozostałe dane stają się mniej widoczne, mniej czytelne będą np. wartości odstające.

#### a) Obliczanie wartości wybranych opisowych charakterystyk rozkładu

Korzystając z danych zawartych w tablicy `recession_table_1` oblicz dla każdego roku następujące opisowe charakterystyki rozkładu:
- wartość średnia (ang. mean),
- odchylenie standardowe (ang. standart deviation),
- kwartyle (ang. quartiles),
- moda (ang. mode).

Zestaw wyniki w tablicy `descriptive_parameters_2`.

Indeks:
 - zawiera następujące nazwy wierszy: mean, std, Q1, Q2, Q3, mode.
 
Kolumny:
 - lata (osobna kolumna dla każdego roku opatrzona nagłówkiem 2006, 2007 itd. sformatowanym jako str)
 
Uwaga! W celu otrzymania maksymalnej liczby punktów otrzymane wyniki powinny być w całości wynikiem pracy napisanego kodu, bez manualnego określania, w którym przedziale znajdują się określone statystyki itp.

```python
# YOUR CODE HERE

# Oświadczamy, że nie zwróciliśmy należytej uwagi na zajęciach oraz niezauważliśmy dołączego pliku
# recession_table_1.pkl, dlatego sami go stworzyliśmy w oparciu o tabelę recession_table_1
# i na tym oparliśmy nasze dalsze działania.

# mean
mean = 0
for column in recession_table_1.columns:
    try:
        mean += column.mid*recession_table_1[column]
    except:
        mean = column.mid*recession_table_1[column]                           
# std
std = 0
for column in recession_table_1.columns:
    try:
        std += (column.mid-mean)**2*recession_table_1[column]
    except:
        std = (column.mid-mean)**2*recession_table_1[column]
std = np.sqrt(std)       

# Q1, Q2, Q3
Q1 = np.zeros(len(recession_table_1))
Q2 = np.zeros(len(recession_table_1))
Q3 = np.zeros(len(recession_table_1))
for index, row in recession_table_1.iterrows():
    selector = recession_table_1.index == index
    for i in range(1,len(recession_table_1.columns)+1):
        if np.sum(row[:i]) < 0.25:
            Q1 = [ Q1[j]+1 if selector[j] == True else Q1[j] for j in range(len(selector)) ]
        if np.sum(row[:i]) < 0.5:
            Q2 = [ Q2[j]+1 if selector[j] == True else Q2[j] for j in range(len(selector)) ]
        if np.sum(row[:i]) < 0.75:
            Q3 = [ Q3[j]+1 if selector[j] == True else Q3[j] for j in range(len(selector)) ]
            
Qq1 = []        
for index, row in recession_table_1.reset_index(drop=True).iterrows():
    l = recession_table_1.columns[int(Q1[index])].left
    hI = recession_table_1.columns[int(Q1[index])].length
    n = int(Q1[index])
    row = row.reset_index(drop=True)
    omega = row[n]
    suma = np.sum(row[:n])
    Qq1.append(l+0.25/omega*hI-suma/omega*hI)
    
Qq2 = []    
for index, row in recession_table_1.reset_index(drop=True).iterrows():
    l = recession_table_1.columns[int(Q2[index])].left
    hI = recession_table_1.columns[int(Q2[index])].length
    n = int(Q2[index])
    row = row.reset_index(drop=True)
    omega = row[n]
    suma = np.sum(row[:n])
    Qq2.append(l+0.5/omega*hI-suma/omega*hI)
    
Qq3 = []
for index, row in recession_table_1.reset_index(drop=True).iterrows():
    l = recession_table_1.columns[int(Q3[index])].left
    hI = recession_table_1.columns[int(Q3[index])].length
    n = int(Q3[index])
    row = row.reset_index(drop=True)
    omega = row[n]
    suma = np.sum(row[:n])
    Qq3.append(l+0.75/omega*hI-suma/omega*hI)
    
Qq1 = pd.DataFrame(Qq1).T
Qq2 = pd.DataFrame(Qq2).T
Qq3 = pd.DataFrame(Qq3).T

# mode
mode = []
for nRow in recession_table_1.index:
    n = np.argmax(recession_table_1.loc[nRow])
    xD = recession_table_1.columns[n].left
    hD = recession_table_1.columns[n].length
    row = recession_table_1.loc[nRow].reset_index(drop=True)
    omega1 = row[n]
    omega2 = row[n-1]
    omega3 = row[n+1]
    mode.append(xD + (omega1-omega2)/((omega1-omega2)+(omega1-omega3))*hD)
mode = pd.DataFrame(mode).T

descriptive_parameters_2 = mean.to_frame().T
descriptive_parameters_2 = pd.concat([descriptive_parameters_2,std.to_frame().T])

Qq1.columns = descriptive_parameters_2.columns
Qq2.columns = descriptive_parameters_2.columns
Qq3.columns = descriptive_parameters_2.columns
descriptive_parameters_2 = pd.concat([descriptive_parameters_2,Qq1,Qq2,Qq3])


mode.columns = descriptive_parameters_2.columns
descriptive_parameters_2 = pd.concat([descriptive_parameters_2,mode])

descriptive_parameters_2.index = ["mean", "std", "Q1", "Q2", "Q3", "mode"]
descriptive_parameters_2.columns.name=""
descriptive_parameters_2.columns = descriptive_parameters_2.columns.astype(str)

display(descriptive_parameters_2)
```

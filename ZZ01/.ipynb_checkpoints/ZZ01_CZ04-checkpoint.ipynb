Na stronie, z której pobrano informacje przy niektórych erupcjach pojawia się +- ileś lat, być może daty erupcji zostały określone arbitralnie (bardzo mało jest wybuchów z pierwszego stycznia, natomiast ponad 2500, czyli prawie 30% nie ma dnia i miesiąca rozpoczęcia). Jest warte zauważanie jest to, że tylko 183 erupcje mają wpisane wartość roku rozpoczęcia i zakończenia bez miesiąca i dnia. Po odrzuceniu eksplozji z brakującymi danymi operujemy zatem tylko na około połowie zbioru danych. Nie możemy jednak wykluczyć, że w niektórych (być może bardzo licznych) przypadkach wpisano datę ropoczęcia i zakończenia w sposób arbitralny. Na podstawie tych danych wyłania nam się obraz, że najdłużej eksplodują wulkany, kiedy VEI jest równe 2 lub 3. W tym zestawieniu nie zostały przedstawione żadne erupcje o indeksie eksplozywności równym 7 (erupcja wulkanu Tambora trwała 3 lata). Widać ogromne zagęszczenie wśród okolic zera. Postanowiliśmy także zbadać histogramy czasu trwania erupcji wulkanów o różnych VEI.

Wyraźnie widać przewagę eksplozji trwających od 0 do 5 dni. Nie dysponujemy dostateczną wiedzą, aby skomentować te wyniki poza tym co już napisaliśmy wyżej o możliwej arbitralności dat. Możemy jeszcze zaobserwować, że wraz ze wzrostem VEI więcej wartości stosunkowo do wszystkich eksplozji o danym indeksie eksplozywności trwa dłużej. Natomiast właśnie pojedyńcze przypadki długiego trwania zaobserwowaliśmy dla VEI równego 2,3 oraz 4. Erupcje wulkanów dla VEI wynoszącego 6 trwały 18,19,153 i 154 dni. Zadziwiające podobieństwo pomiędzy tymi czasami, parami różnią się o 1 dzień. Analiza czasu trwania eksplozji dostarczyła na pewnych informacji poglądowych, dowiedzieliśmy się, że erupcje mogą trwać bardzo długo, natomaist uważamy, że dane jednak nie są odpowiednie do głębszych rozważań na ten temat.

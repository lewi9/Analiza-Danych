#### a) Wizualizacja

Zwizualizuj funkcje gęstości prawdopodobieństwa następujących rozkładów teoretycznych:
- Wykres 1: $N(0, 1)$, $N(0, 0.75)$, $N(0, 1.25)$, $N(-2, 1)$, $N(2, 1)$
- Wykres 2: $N(0, 1)$, $t(1)$, $t(2)$, $t(5)$, $t(30)$
- Wykres 3: $\chi^2(1)$, $\chi^2(2)$, $\chi^2(3)$, $\chi^2(5)$, $\chi^2(8)$
- Wykres 4: $F(10, 1)$, $F(10, 2)$, $F(10, 5)$, $F(10, 10)$, $F(10, 20)$

- $\chi^2(df)$ - rozkład chi-kwadrat o $df$ stopniach swobody,

```python
# YOUR CODE HERE

N = [(0,1), (0,0.75), (0,1.25), (-2,1), (2,1)]
T = [1, 2, 5, 30]
X = [1, 2, 3, 5, 8]
F1 = [(10,1), (10,2), (10,5), (10,10), (10,20)]
F2 = [(1,10), (2,10), (5,10), (10,10), (20,10)]
F3 = [(1,1), (2,2), (5,5), (25,25), (50,50)]

title = ["Normal distribution", "T-student distribution", "Chi-2 distribution", 
         "F Snedocer distribution (1)", "F Snedocer distribution (2)", "F Snedocer distribution (3)"]

colors = ['r', 'g', 'b', 'c', 'm', 'y']
fig, axs = plt.subplots(2,3,figsize=(15,10), facecolor='white')


x1 = np.linspace(-5,5,1000)
x2 = np.linspace(0,5,500)

for i, elem in enumerate(N):
    sns.lineplot(x=x1, y=sp.stats.norm.pdf(x1, elem[0], elem[1]), color=colors[i], ax=axs[0,0], label=f"N({elem[0]}, {elem[1]})")

sns.lineplot(x=x1, y=sp.stats.norm.pdf(x1, 0,1), color=colors[0], ax=axs[0,1], label="N(0,1)")
for i, elem in enumerate(T):
    sns.lineplot(x=x1, y=sp.stats.t.pdf(x1, elem), color=colors[i+1], ax=axs[0,1], label=f"t({elem})")

for i, elem in enumerate(X):
    sns.lineplot(x=x2, y=sp.stats.chi2.pdf(x2, elem), color=colors[i], ax=axs[0,2], label=f"chi2({elem})")
    
for i, elem in enumerate(F1):
    sns.lineplot(x=x2, y=sp.stats.f.pdf(x2, elem[0], elem[1]), color=colors[i], ax=axs[1,0], label=f"F({elem[0]}, {elem[1]})")
    
for i, elem in enumerate(F2):
    sns.lineplot(x=x2, y=sp.stats.f.pdf(x2, elem[0], elem[1]), color=colors[i], ax=axs[1,1], label=f"F({elem[0]}, {elem[1]})")
    
for i, elem in enumerate(F3):
    sns.lineplot(x=x2, y=sp.stats.f.pdf(x2, elem[0], elem[1]), color=colors[i], ax=axs[1,2], label=f"F({elem[0]}, {elem[1]})")
    
for i, ax in enumerate(axs.flatten()):
    if i < 2:
        ax.set_xlim([-5, 5])
        ax.set_ylim([0, 0.8])
    else:
        ax.set_xlim([0, 5])
        ax.set_ylim([0, 1.6])
    ax.set_facecolor("white")
    ax.set_title(title[i])
    ax.set_xlabel("Value")
    ax.set_ylabel("Probability density")
    ax.legend(loc = "upper right", title="Distribution")
    
axs[0,2].set_ylim([0, 0.8])
    
fig.suptitle("Different distributions")
fig.tight_layout()
```

#### b) Charakterystyka rozkładów
Na podstawie przygotowanych wykresów scharakteryzuje opisane rozkłady teoretyczne i odpowiedz na pytanie w jaki sposób zmiana wartości parametrów rozkładów wpływa na położenie i kształt ich funkcji gęstości prawdopodobieństwa.

**Rozkład normalny** - posiada charakterystyczny dla siebie kształt (w kształcie dzwonu), zmieniając średnią zmieniamy jego położenie względem osi x, natomiast odchylenie standardowe wpływa na kształt poprzez kurtozę. Ta dystrybucja ma najwyżej położony punkt dla wartości średniej. Wraz ze wzrostem std, staje się on niższy i bardziej rozciągnięty na boki. Jest symetryczny względem średniej.

**Rozkład t-studenta** - posiada podobny kształt do rozkładu normalnego. Posiada on jednak "cięższe ogony", przez co wolniej zbiega do bardzo małych wartości niż rozkład normalny. W przypadku wzrostu liczby stopni swobody upodabnia się on coraz bardziej do rozkładu normalnego. Jest symetryczny względem zera.

**Rozkład chi-kwadrat** - posiada on inny kształt niż jego poprzednicy, przede wszystkim jest on prawostronnie asymetryczny, nie posiada ujemnych wartości (co wynika z jego specyfiki, ponieważ suma kwadratów jest nieujemna), wartość średnia jest równa liczbie stopni swobody. Wraz ze wzrostem liczby stopni swobody, jego skośność maleje, zmienia się też kurtoza. Też ma "ciężkie ogony".

**Rozkład F-Snedecora** - Również nie posiada wartości ujemnych, skośny, prawostronnie asymetryczny, z "ciężkimi ogonami". Dąży do rozkładu normalnego wraz ze wzrostem obu stopni swobody. Jest on stosunkiem dwóch zmiennych losowych o zadanych stopniach swobody, dlatego wraz ze wzrostem jednego stopnia swobody przesuwa się w prawo oraz zmienia się jego wysokość. Czerwone i zielone linie odbiegają od reszty, tak się dzieje gdy pierwszy stopień swobody jest równy jeden bądź dwa. Jeżeli drugi stopień swobody jest równy 1 lub 2 to taka sytuacja nie zachodzi.

```python
# p1
# YOUR CODE HERE
p1 = sp.stats.t.cdf(-1.5, 10)

distribution = "t"
parameters = ["10"]
height = 0.8
x0 = np.linspace(-20, -1.5, 1000)

title = (f"P(X < -1.5) = {np.round(p1, 4)}")

xl_parameters = parameters.copy()
xl_parameters.insert(0, "0.0001")

xr_parameters = parameters.copy()
xr_parameters.insert(0, "0.9999")

x_l = eval(f"sp.stats.{distribution}.ppf({','.join(xl_parameters)})")
x_r = eval(f"sp.stats.{distribution}.ppf({','.join(xr_parameters)})")
x = np.linspace(x_l, x_r, 1000)

plot_parameters = parameters.copy()
plot_parameters.insert(0, "x")
y = eval(f"sp.stats.{distribution}.pdf({','.join(plot_parameters)})")

plot_parameters[0] = "x0"
y0 = eval(f"sp.stats.{distribution}.pdf({','.join(plot_parameters)})")
yt0 = [0 for i in range(len(y0))]

fig, ax = plt.subplots(1, 1, figsize=(5,5), facecolor="white")
sns.lineplot(x=x, y=y, ax=ax, color="black")
ax.fill_between(x0, yt0, y0, color="green")
ax.set_ylim([0,height])
ax.set_xlim([x_l, x_r])
ax.set_facecolor("white")
ax.set_title(title)
ax.set_xlabel("x")
ax.set_ylabel("Probability density")
```

```python
# s1
# YOUR CODE HERE
# P(X < s) = 0.25
s1 = sp.stats.chi2.ppf(0.25, 5)

s = s1
distribution = "chi2"
parameters = ["5"]
height = 0.8
x0 = np.linspace(0, s, 1000)

title = (f"P(X < {np.round(s1, 4)}) = 0.25")

xl_parameters = parameters.copy()
xl_parameters.insert(0, "0.0001")

xr_parameters = parameters.copy()
xr_parameters.insert(0, "0.9999")

x_l = eval(f"sp.stats.{distribution}.ppf({','.join(xl_parameters)})")
x_r = eval(f"sp.stats.{distribution}.ppf({','.join(xr_parameters)})")
x = np.linspace(x_l, x_r, 1000)

plot_parameters = parameters.copy()
plot_parameters.insert(0, "x")
y = eval(f"sp.stats.{distribution}.pdf({','.join(plot_parameters)})")

plot_parameters[0] = "x0"
y0 = eval(f"sp.stats.{distribution}.pdf({','.join(plot_parameters)})")
yt0 = [0 for i in range(len(y0))]

fig, ax = plt.subplots(1, 1, figsize=(5,5), facecolor="white")
sns.lineplot(x=x, y=y, ax=ax, color="black")

vline_parameters = parameters.copy()
vline_parameters.insert(0, str(s))

ax.vlines(x = s,
          ymin = 0,
          ymax = eval(f"sp.stats.{distribution}.pdf({','.join(vline_parameters)})"),
          colors = 'red',
          label = 'vline_multiple - full height')
ax.fill_between(x0, yt0, y0, color="green")
ax.set_ylim([0,height])
ax.set_xlim([x_l, x_r])
ax.set_facecolor("white")
ax.set_title(title)
ax.set_xlabel("x")
ax.set_ylabel("Probability density")
```

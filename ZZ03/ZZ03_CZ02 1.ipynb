```python
# Agregacja danych z poszczególnych dni roku
# YOUR CODE HERE
daily_bike_traffic_data = bike_traffic_data.copy().groupby(pd.Grouper(freq='1d')).sum()
daily_bike_traffic_data.to_pickle("daily_bike_traffic_data.pkl")
daily_bike_traffic_data = pd.read_pickle("daily_bike_traffic_data.pkl")
display(daily_bike_traffic_data.head(10).style.format(precision=0))
```

```python
# Agregacja danych z poszczególnych dni tygodnia
# YOUR CODE HERE
bike_traffic_by_weekday_data = daily_bike_traffic_data.copy()
bike_traffic_by_weekday_data["Day"] = bike_traffic_by_weekday_data.index.astype(str)
bike_traffic_by_weekday_data["Day"] = bike_traffic_by_weekday_data["Day"].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d').strftime('%A'))
bike_traffic_by_weekday_data = bike_traffic_by_weekday_data.groupby("Day").mean()
bike_traffic_by_weekday_data = bike_traffic_by_weekday_data.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
display(bike_traffic_by_weekday_data.style.format(precision=2))
```

```python
# Agregacja danych z konkretnych okresów tygodnia (dzień roboczy/weekend)
# YOUR CODE HERE
weekday_weekend_hourly_bike_traffic_data = bike_traffic_data.copy()
weekday_weekend_hourly_bike_traffic_data["Hour"] = weekday_weekend_hourly_bike_traffic_data.index.astype(str)
weekday_weekend_hourly_bike_traffic_data["Hour"] = weekday_weekend_hourly_bike_traffic_data["Hour"].apply(lambda x: int(datetime.datetime.strptime(x, '%Y-%m-%d %H:%M').strftime('%H')))
weekday_weekend_hourly_bike_traffic_data["Day"] = weekday_weekend_hourly_bike_traffic_data.index.astype(str)
weekday_weekend_hourly_bike_traffic_data["Day"] = weekday_weekend_hourly_bike_traffic_data["Day"].apply(lambda x: int(datetime.datetime.strptime(x, '%Y-%m-%d %H:%M').strftime('%w')))


df1 = weekday_weekend_hourly_bike_traffic_data[(weekday_weekend_hourly_bike_traffic_data["Day"] < 6) & (weekday_weekend_hourly_bike_traffic_data["Day"] > 0) ].drop("Day", axis=1)
df2 = weekday_weekend_hourly_bike_traffic_data[(weekday_weekend_hourly_bike_traffic_data["Day"] == 0) | (weekday_weekend_hourly_bike_traffic_data["Day"] == 6)].drop("Day", axis=1)
weekday_weekend_hourly_bike_traffic_data = pd.concat([df1.groupby("Hour").mean(), df2.groupby("Hour").mean()], axis=1)

tuples = [("Weekday","Total"),("Weekday","East"),("Weekday","West"),("Weekend","Total"),("Weekend","East"),("Weekend","West")]
columns = pd.MultiIndex.from_tuples(tuples, names=["1 poziom", "2 poziom"])
weekday_weekend_hourly_bike_traffic_data.columns = columns
display(weekday_weekend_hourly_bike_traffic_data.style.format(precision=2))
weekday_weekend_hourly_bike_traffic_data.to_pickle("weekday_weekend_hourly_bike_traffic_data.pkl")
```

Przygotuj tablicę `monthly_weather_data` zawierającą dane z tablicy `weather_data` zagregowane w obrębie poszczególnych miesięcy według następującego klucza:

- wartość maksymalna z całego miesiąca: Temperature: Max, Wind: Max speed,
- suma z całego miesiąca: Precipitation: Total, Snow: Fall,

Tablica powinna zawierać wyłącznie następujące elementy:</br>
> Indeks: Date</br>
> PeriodIndex zawierający okresy czasu, których dotyczą dane z raportów pogodowych.</br>

> Kolumny: W podanej niżej kolejności</br>
> Temperature: Min, Avg, Max</br>
> Precipitation: Total</br>
> Snow: Fall, Depth</br>
> Wind: Avg speed, Max speed</br>
> Sky: Cloud cover</br>
> (Nazwa 1 poziomu: podległe jej nazwy 2 poziomu)

Wyświetl przygotowaną tablicę w taki sposób aby zawarte w nich wartości wyświetlały się z dokładnością 2 miejsc po przecinku.

```python
# YOUR CODE HERE
weather_data[("Snow","Depth")] = weather_data[("Snow","Depth")].fillna(0)

tuples = [("Temperature", "Min"), ("Temperature","Avg"), ("Temperature", "Max"), ("Precipitation","Total"), ("Snow","Fall"),
           ("Snow","Depth"), ("Wind","Avg speed"), ("Wind", "Max speed"), ("Sky", "Cloud cover")]
columns = pd.MultiIndex.from_tuples(tuples, names=["1 poziom", "2 poziom"])
agg = weather_data.copy().groupby(pd.Grouper(freq='M'))
operations = ['min','mean','max','sum','sum','mean','mean','max','mean']

monthly_weather_data = pd.DataFrame()
for i in range(len(columns)):
    monthly_weather_data[columns[i:i+1]] = agg[columns[i:i+1]].agg(operations[i])
monthly_weather_data.columns = columns
display(monthly_weather_data.style.format(precision=2))
```

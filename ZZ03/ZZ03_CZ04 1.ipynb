```python
# Miejsce na dodatkowy kod

#==============================================================================================================
# Odszumianie "dni roboczych" z dni wolnych od pracy

bike_traffic_data = pd.read_pickle("bike_traffic_data.pkl")
pivoted = bike_traffic_data.pivot_table(['Total'],
                           index=bike_traffic_data.index.dayofyear,
                           columns=bike_traffic_data.index.hour,
                           fill_value=0)
pivoted.head()

# Normalizujemy dane, dla każdego dnia, aby zbadać gdzie jest najmniejsze natężenie, a gdzie największe
# Nie interesuje nas ile w danym dniu przejechało rowerzystów, tylko jak rozkładało się to w czasie

mins = np.min(pivoted, axis=1)
maxs = np.max(pivoted, axis=1)

pivoted = pivoted.sub(mins, axis=0).div(maxs-mins, axis=0)


## T-SNE - redukcja wymiarowości w celu wizualizacji

tsne = sklearn.manifold.TSNE(n_components=2, perplexity=16, random_state=42)
result = tsne.fit_transform(pivoted)

## Ze względu na wielowymiarowość klastrów została użyta klasteryzacja Gaussian Mixture Model
## Metoda T-SNE jest tylko poglądowa, można manipulować wizualizacją za pomocą parametru perplexity

GMM = sklearn.mixture.GaussianMixture(n_components=2, random_state=42)
predicted = GMM.fit_predict(pivoted)

fig, axs = plt.subplots(1,1, figsize=(8,8), facecolor="white")
axs.scatter(result[:,0], result[:,1], c=predicted)
axs.set_title("Wizualizacja danych - badanie natężenia ruchu w zależności od godziny.\nTsne, wizualizacja w przestrzeni dwuwymiarowej")
axs.set_xlabel("Składowa 1")
axs.set_ylabel("Składowa 2")
axs.set_facecolor("white")

fig.tight_layout()
plt.savefig("Figure_7.png")

daily_bike_traffic_data = pd.read_pickle("daily_bike_traffic_data.pkl")
daily_bike_traffic_data["cluster"] = predicted
daily_bike_traffic_data["weekDay"] = daily_bike_traffic_data.index.day_of_week

new_predicted = [x  for x in predicted for i in range(24)]
bike_traffic_data["cluster"] = new_predicted
bike_traffic_data["Hour"] = bike_traffic_data .index.astype(str)
bike_traffic_data["Hour"] = bike_traffic_data ["Hour"].apply(lambda x: int(datetime.datetime.strptime(x, '%Y-%m-%d %H:%M').strftime('%H')))

df1 = bike_traffic_data[bike_traffic_data.cluster == 0 ].drop("cluster", axis=1)
df2 = bike_traffic_data[bike_traffic_data.cluster == 1 ].drop("cluster", axis=1)
weekday_weekend_hourly_bike_traffic_data = pd.concat([df1.groupby("Hour").mean(), df2.groupby("Hour").mean()], axis=1)

tuples = [("Weekday","Total"),("Weekday","East"),("Weekday","West"),("Weekend","Total"),("Weekend","East"),("Weekend","West")]
columns = pd.MultiIndex.from_tuples(tuples)
weekday_weekend_hourly_bike_traffic_data.columns = columns
#display(weekday_weekend_hourly_bike_traffic_data.style.format(precision=2))

dfStyled = weekday_weekend_hourly_bike_traffic_data.style.format(precision=2)
dfi.export(dfStyled, "bike_traffic_removed_noise_3.png", chrome_path=None, fontsize=10)

fig, axs = plt.subplots(1,3, figsize=(15,5), facecolor="white")

axs[0].set_facecolor("white")
axs[0].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekday","Total")], c="purple", alpha=0.8)
axs[0].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekday","East")], c="green", alpha=0.8)
axs[0].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekday","West")], c="orange", alpha=0.8)

axs[0].set_xlim([0,23])
axs[0].set_ylim([0,350])

axs[0].set_xlabel("godzina dnia")
axs[0].set_ylabel("średnia liczba rowerzystów")
axs[0].set_title("Średnia liczba rowerzystów \no poszczególnych godzinach w dni robocze")

axs[0].xaxis.set_major_locator(mpl.ticker.AutoLocator())
axs[0].xaxis.set_minor_locator(mpl.ticker.AutoMinorLocator())

axs[1].set_facecolor("white")
axs[1].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekend","Total")], c="purple", alpha=0.8, label="Total")
axs[1].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekend","East")], c="green", alpha=0.8, label="East")
axs[1].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekend","West")], c="orange", alpha=0.8, label="West")

axs[1].set_xlim([0,23])
axs[1].set_ylim([0,350])

axs[1].set_xlabel("godzina dnia")
axs[1].set_ylabel("średnia liczba rowerzystów")
axs[1].set_title("Średnia liczba rowerzystów \no poszczególnych godzinach w dni 'wolne'")

axs[1].xaxis.set_major_locator(mpl.ticker.AutoLocator())
axs[1].xaxis.set_minor_locator(mpl.ticker.AutoMinorLocator())

leg = axs[1].legend(loc='center', bbox_to_anchor=(0.5, -0.30), shadow=False, ncols=3)

axs[2].set_facecolor("white")
sns.barplot(x=sorted(daily_bike_traffic_data[daily_bike_traffic_data.cluster == 1].weekDay.unique()), 
            y=daily_bike_traffic_data[daily_bike_traffic_data.cluster == 1].groupby("weekDay").size(),
            ax=axs[2], color="magenta")
axs[2].set_xticklabels(["Monday","Tuesday","Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], rotation=45)
axs[2].set_xlabel("dzień tygodnia")
axs[2].set_ylabel("wystąpienia")
axs[2].set_title("Liczba dni tygodnia jako dni 'wolne'")

fig.suptitle("Rowerzyści na moście Fremont Bridge w Seattle")
fig.tight_layout()

plt.savefig("Figure_8.png")

old_data = pd.read_pickle("weekday_weekend_hourly_bike_traffic_data.pkl")

fig, axs = plt.subplots(1,2, figsize=(10,5), facecolor="white")

axs[0].set_facecolor("white")
axs[0].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekday","Total")], c="purple", alpha=0.8)
axs[0].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekday","East")], c="green", alpha=0.8)
axs[0].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekday","West")], c="orange", alpha=0.8)
axs[0].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), old_data[("Weekday","Total")], c="pink", alpha=0.8)
axs[0].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), old_data[("Weekday","East")], c="chartreuse", alpha=0.8)
axs[0].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), old_data[("Weekday","West")], c="wheat", alpha=0.8)

axs[0].set_xlim([0,23])
axs[0].set_ylim([0,350])

axs[0].set_xlabel("godzina dnia")
axs[0].set_ylabel("średnia liczba rowerzystów")
axs[0].set_title("Średnia liczba rowerzystów \no poszczególnych godzinach w dni robocze")

axs[0].xaxis.set_major_locator(mpl.ticker.AutoLocator())
axs[0].xaxis.set_minor_locator(mpl.ticker.AutoMinorLocator())

axs[1].set_facecolor("white")
axs[1].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekend","Total")], c="purple", alpha=0.8, label="Total")
axs[1].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekend","East")], c="green", alpha=0.8, label="East")
axs[1].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), weekday_weekend_hourly_bike_traffic_data[("Weekend","West")], c="orange", alpha=0.8, label="West")
axs[1].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), old_data[("Weekend","Total")], c="pink", alpha=0.8, label="Total_old")
axs[1].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), old_data[("Weekend","East")], c="chartreuse", alpha=0.8, label='East_old')
axs[1].plot(weekday_weekend_hourly_bike_traffic_data.index.astype(str), old_data[("Weekend","West")], c="wheat", alpha=0.8, label="West_old")


axs[1].set_xlim([0,23])
axs[1].set_ylim([0,350])

axs[1].set_xlabel("godzina dnia")
axs[1].set_ylabel("średnia liczba rowerzystów")
axs[1].set_title("Średnia liczba rowerzystów \no poszczególnych godzinach w dni 'wolne'")

axs[1].xaxis.set_major_locator(mpl.ticker.AutoLocator())
axs[1].xaxis.set_minor_locator(mpl.ticker.AutoMinorLocator())

fig.legend(loc='center', bbox_to_anchor=(0.5, -0.05), shadow=False, ncols=6)

fig.suptitle("Rowerzyści na moście Fremont Bridge w Seattle")
fig.tight_layout()

plt.savefig("Figure_9.png", bbox_inches='tight')

#==============================================================================================================
# Korelacja spearmana

daily_bike_traffic_data = pd.read_pickle("daily_bike_traffic_data.pkl")
weather_data = pd.read_pickle("weather_data.pkl")


df = pd.concat([weather_data, daily_bike_traffic_data], axis=1)
R_corr_1 = df.corr("pearson")
R_corr_1 = R_corr_1.loc[weather_data.columns, daily_bike_traffic_data.columns]
R_corr_1.index = weather_data.columns

df = pd.concat([weather_data, daily_bike_traffic_data], axis=1)
R2_corr_1 = df.corr("spearman")
R2_corr_1 = R2_corr_1.loc[weather_data.columns, daily_bike_traffic_data.columns]
R2_corr_1.index = weather_data.columns

fig, axs = plt.subplots(1,2, figsize=(15,8), facecolor="white")

axs[0].set_facecolor("white")
sns.heatmap(R_corr_1, ax=axs[0], cmap="coolwarm", annot=True, vmin=-1, vmax=1, fmt=".3f")
axs[0].set_xlabel("Przejazdy przez most")
axs[0].set_ylabel("Parametry pogody")
axs[0].set_title("Współczynnik liniowej korelacji pearsona")


axs[1].set_facecolor("white")
sns.heatmap(R2_corr_1, ax=axs[1], cmap="coolwarm", annot=True, vmin=-1, vmax=1, fmt=".3f")
axs[1].set_xlabel("Przejazdy przez most")
axs[1].set_ylabel("Parametry pogody")
axs[1].set_title("Współczynnik rangowej korelacji spearmana")

fig.suptitle("Relacja między ilością rowerzystów, a pogodą")
fig.tight_layout()

plt.savefig("Figure_10.png")

R_corr_2 = weather_data.corr("pearson")
R2_corr_2 = weather_data.corr("spearman")

fig, axs = plt.subplots(1,2, figsize=(15,8), facecolor="white")

axs[0].set_facecolor("white")
sns.heatmap(R_corr_2, ax=axs[0], cmap="coolwarm", annot=True, vmin=-1, vmax=1, fmt=".3f")
axs[0].set_xlabel("Parametry pogody")
axs[0].set_ylabel("Parametry pogody")
axs[0].set_title("Współczynnik liniowej korelacji pearsona")


axs[1].set_facecolor("white")
sns.heatmap(R2_corr_2, ax=axs[1], cmap="coolwarm", annot=True, vmin=-1, vmax=1, fmt=".3f")
axs[1].set_xlabel("Parametry pogody")
axs[1].set_ylabel("Parametry pogody")
axs[1].set_title("Współczynnik rangowej korelacji spearmana")

fig.suptitle("Relacja między parametrami pogody")
fig.tight_layout()

plt.savefig("Figure_11.png")

#==============================================================================================================
# Badanie różnicy pomiędzy ludźmi, którzy przejeżdżali stronami mostu, a pogodą
daily_bike_traffic_data = pd.read_pickle("daily_bike_traffic_data.pkl")
weather_data = pd.read_pickle("weather_data.pkl")

df1 = pd.DataFrame()

df1["difference"] = abs(daily_bike_traffic_data["West"]/daily_bike_traffic_data["Total"] 
                        - daily_bike_traffic_data["East"]/daily_bike_traffic_data["Total"])
df = pd.concat([weather_data, df1], axis=1)

columns = pd.MultiIndex.from_tuples([("Pearson","Difference"), ("Spearman","Difference")])

R_corr_1 = df.corr("pearson")
R_corr_1 = R_corr_1.loc[weather_data.columns, df1.columns]
R_corr_1.index = weather_data.columns

R2_corr_1 = df.corr("spearman")
R2_corr_1 = R2_corr_1.loc[weather_data.columns, df1.columns]
R2_corr_1.index = weather_data.columns

result = pd.concat([R_corr_1, R2_corr_1], axis=1)
result.columns = columns

dfStyled = result.style.format(precision=3)
dfi.export(dfStyled, "corr_5.png", chrome_path=None)

#==============================================================================================================
# Badanie zależności między dniem roku, a liczbą rowerzystów
daily_bike_traffic_data = pd.read_pickle("daily_bike_traffic_data.pkl")

day = daily_bike_traffic_data.index.dayofyear
def func(params):
    a, b, c = params
    if(b==0):
        b = np.inf
    b *= 365
    return a*np.sin(data_x/b*2*np.pi) + c

def con_func(params):
    a, b, c = params
    b *= 365
    return b-int(b)

constrains = ({'type': 'eq', 'fun': lambda coeffs: con_func(coeffs)})
    
def func_to_minimise(params):
    y_pred = func(params)
    return np.sum((y_pred - data_y) ** 2)

factors=[]
for column_bike in daily_bike_traffic_data:
    mask = ~np.isnan(daily_bike_traffic_data[column_bike])
    data_x = day[mask]
    data_y = daily_bike_traffic_data[column_bike][mask]
    options = {'maxiter': 10000}
    res = sp.optimize.minimize( func_to_minimise, x0=[0,2,0], constraints=constrains, options=options)
    factors.append(res.x)
    
fig, axs = plt.subplots(3,1, figsize=(15,10), facecolor="white")
for i in range(len(daily_bike_traffic_data.columns)):
    axs[i].set_facecolor("white")
    axs[i].set_title(f"{daily_bike_traffic_data.columns[i]}, równanie regresji: y = {factors[i][0]:.3f}*sin(x/{factors[i][1]*365}*2pi) + {factors[i][2]:.3f}")
    axs[i].set_xlabel("Day of year")
    axs[i].set_ylabel("Bikers")
    axs[i].scatter(day, daily_bike_traffic_data.iloc[:,i], c="black", alpha=0.6)
    x = np.arange(1,366,1)
    data_x = day[mask]
    axs[i].plot(x, func(factors[i]), c="red", linewidth=5)
    axs[i].set_ylim([0,5000])
    axs[i].set_xlim([1,365])
fig.suptitle("Wykresy korelacyjne dnia roku z liczbą rowerzystów\n")
fig.tight_layout()

plt.savefig("Figure_12.png")

#==============================================================================================================
# Regresja wieloraka
daily_bike_traffic_data = pd.read_pickle("daily_bike_traffic_data.pkl")
weather_data = pd.read_pickle("weather_data.pkl")

day = daily_bike_traffic_data.index.dayofyear

columns = [("Temperature","Avg"), ("Sky", "Cloud cover"), ("Wind","Max speed")]

def func1(a):
    return a*weather_data[columns[0]]

def func2(a,b):
    return a*np.exp(-b*weather_data[columns[1]]) 

def func3(a):
    return a*weather_data[columns[2]]

def func4(a):
    return a*np.sin(day/720*2*np.pi)

def func(params):
    a1, a2, b2, a3, a4, c = params
    return func1(a1) + func2(a2,b2) + func3(a3) + func4(a4) + c

def func_to_minimise(params):
    y_pred = func(params)
    return np.sum((y_pred - daily_bike_traffic_data["Total"]) ** 2)

mask = (~np.isnan(daily_bike_traffic_data["Total"])
        & ~np.isnan(weather_data[columns[0]])
        & ~np.isnan(weather_data[columns[1]])
        & ~np.isnan(weather_data[columns[2]]))

day = day[mask]
weather_data = weather_data[mask]
daily_bike_traffic_data = daily_bike_traffic_data[mask]

options = {'maxiter': 10000}
res = sp.optimize.minimize( func_to_minimise, x0=np.zeros(6), options=options)

fig, axs = plt.subplots(1,1, figsize=(15,5), facecolor="white")

axs.set_facecolor("white")
axs.set_title(f"Regresja wieloraka, użyte kolumny {columns}, day_of_year")
axs.set_xlabel("Day of year")
axs.set_ylabel("Total bikers")
axs.scatter(day, daily_bike_traffic_data.iloc[:,0], c="black", alpha=0.6)
x = np.arange(1,366,1)
x = x[mask]
axs.plot(x, func(res.x), c="red", linewidth=5, label="Regresja wieloraka")

axs.set_ylim([0,5000])
axs.set_xlim([1,365])

MAE = abs(daily_bike_traffic_data["Total"][mask]-func(res.x)).mean()
R2 = 1 - (np.sum(np.square(daily_bike_traffic_data["Total"][mask]-func(res.x)))
          /np.sum(np.square(daily_bike_traffic_data["Total"][mask]
                            -daily_bike_traffic_data["Total"][mask].mean())))

labels=[f'y = {res.x[0]:.3f}*{columns[0]}', f"y = {res.x[1]:.3f}*exp({-res.x[2]:.3f}*{columns[1]})",
        f"y = {res.x[3]:.3f}*{columns[2]}", f"y = {res.x[4]:.3f}*sin(day*pi/365) + {res.x[5]:.3f}",
        f"MAE = {MAE:.3f}", f"R2 = {R2}"]

handles = [mpl.lines.Line2D([], [], color='white', label=x) for x in labels]

fig.legend(handles, labels, ncols=3, loc='center', bbox_to_anchor=(0.5, -0.05))

fig.tight_layout()
plt.savefig("Figure_13.png", bbox_inches='tight')

#==============================================================================================================
# Anomalie
daily_bike_traffic_data = pd.read_pickle("daily_bike_traffic_data.pkl")
weather_data = pd.read_pickle("weather_data.pkl")

april = daily_bike_traffic_data[daily_bike_traffic_data.index.month == 4]
april = april[april.index.day < 15]
weather_april = weather_data.loc[april.index,:]

df1 = april.copy()
df1[weather_april.columns] = weather_april
#display(df1)

july = daily_bike_traffic_data[daily_bike_traffic_data.index.month == 7]
july = july[july.index.day < 6]
weather_july = weather_data.loc[july.index,:]

df2 = july.copy()
df2[weather_july.columns] = weather_july
#display(df2)

dfStyled = df1.style.format(precision=2)
dfi.export(dfStyled, "anomaly_1.png", chrome_path=None, fontsize=10)

dfStyled = df2.style.format(precision=2)
dfi.export(dfStyled, "anomaly_2.png", chrome_path=None, fontsize=10)

bike_traffic_data = pd.read_pickle("bike_traffic_data.pkl")

april7 = bike_traffic_data[(bike_traffic_data.index.day==7) & (bike_traffic_data.index.month==4)]
april7.index = april7.index.hour

july3 = bike_traffic_data[(bike_traffic_data.index.day==3) & (bike_traffic_data.index.month==7)]
july3.index = july3.index.hour

fig, ax = plt.subplots(2,1, figsize=(8,8), facecolor="white")
ax[0].set_facecolor("white")
ax[0].plot(april7.index, april7.Total, c="purple")
ax[0].set_xlim([0,23])
ax[0].set_ylim([0,700])
ax[0].set_xlabel("Godzina dnia")
ax[0].set_ylabel("Liczba rowerzystów (Total)")
ax[0].set_title("Rowerzyści na Fremont Bridge - 7 kwietnia 2022")

ax[1].set_facecolor("white")
ax[1].plot(july3.index, july3.Total, c="purple")
ax[1].set_xlim([0,23])
ax[1].set_ylim([0,700])
ax[1].set_xlabel("Godzina dnia")
ax[1].set_ylabel("Liczba rowerzystów (Total)")
ax[1].set_title("Rowerzyści na Fremont Bridge - 3 lipca 2022")

fig.suptitle("Anomalie")
fig.tight_layout()

plt.savefig("Figure_14")
```

Nie odrzuciliśmy również hipotezy, że może to być działanie w pełni losowe, rowerzyści jadący po chodniku mogą czasem zignorować obowiązujący ruch prawostronny.

W przypadku wiatru zaskoczyliśmy się, że silniejsza jest korelacja liniowa niż rangowa. Nie wpływa ona tak znącząco jak opady, czy temperatura, natomiast ma ona niewielki wpływ na ruch rowerzystów. Wraz ze wzrostem siły wiatru, ruch rowerowy maleje.

Zachmurzenie także jest istotne dla ruchu rowerowego. Tutaj wiemy, że silne zachmurzenie wiąże się ze spadkiem temperatury oraz wzrostem szans na opady, co może bezpośrednio przyczyniać się do zmniejszenia intensywności ruchu na moście Fremont Bridge.
